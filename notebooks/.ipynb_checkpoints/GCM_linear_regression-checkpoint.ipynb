# Linear regression

In order to test the hypothesized correlation between the ranks of GCMs and the prototypicality of an entity in a category, linear regression was performed and verified against the gold standard. The [statsmodels](https://www.statsmodels.org/stable/index.html) library was used for this. The results can be found in the Jupyter notebook and in the thesis. Plots for every category are saved on GitHub.

```python
# Function for standardized scaling and normalisation of GCM values
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

def scaler(data_frame):
    helper_df = data_frame.copy()
    to_scale = preprocessing.StandardScaler()

    helper_df[['degree', 'eigenvector', 'pagerank']] = to_scale.fit_transform(helper_df[['degree', 'eigenvector', 'pagerank']])
    return helper_df

#df_ig_gcm = scaler(df_ig_gcm)
```

```python
# Function for merging the sparql category with GCM data

# Merge the queried category df to the GCM metrics based on <resource> with a left outer join
# (Keep every row in the category dataframe -> missing values get NaN)

def catMergerNorm(sparql_data, gcm_data):
    df_sparql = sparql_data.copy()
    df_gcm = gcm_data.copy()
    helper_df = pd.merge(df_sparql,
                 df_gcm[['resource','degree', 'eigenvector', 'pagerank']],
                 on='resource', 
                 how='left')
    
    # scale and normalize the gcm metrics (same as scaler function)
    to_scale = preprocessing.StandardScaler()
    helper_df[['degree', 'eigenvector', 'pagerank']] = to_scale.fit_transform(helper_df[['degree', 'eigenvector', 'pagerank']])
    
    print("Dataframe merged with shape: {}".format(helper_df.shape))
    print('============================')

    print("Unmatched rows: ", helper_df.shape[0] - helper_df.dropna().shape[0]) #return amount of rows containing NaN values
    
    return helper_df

def catMerger(sparql_data, gcm_data):
    df_sparql = sparql_data.copy()
    df_gcm = gcm_data.copy()
    helper_df = pd.merge(df_sparql,
                 #df_gcm[['resource','degree','pagerank']],
                 df_gcm[['resource','degree', 'eigenvector', 'pagerank']],
                 on='resource', 
                 how='left')
    
    # scale and normalize the gcm metrics (same as scaler function)
    #to_scale = preprocessing.StandardScaler()
    #helper_df[['degree', 'eigenvector', 'pagerank']] = to_scale.fit_transform(helper_df[['degree', 'eigenvector', 'pagerank']])
    
    print("Dataframe merged with shape: {}".format(helper_df.shape))
    print('============================')

    #print(helper_df.shape[0] - helper_df.dropna().shape[0]) #return amount of rows containing NaN values
    
    return helper_df
```

```python
# Function for merging the survey category data with the gcm category data
  # Introduces "amount_rank", a rank based on the amount of answers for a category compared to the sum of answers in total

# Merge the survey category df to the GCM metrics per category based on <resource> with a left outer join
# (Keep every row in the category dataframe -> missing values get NaN)

def prepSurGCMLinReg(survey_data, cat_gcm_data):
    df_survey = survey_data.copy()
    df_cat_gcm = cat_gcm_data.copy()
    df_survey.rename(columns={"uri": "resource"}, inplace=True) #rename column to match GCM dataframe
        
    # Count how often an answer was given in the survey and save into new dataframe
    survey_count = df_survey['resource'].value_counts() #normalize values
    df_survey_counted = pd.DataFrame({'resource':survey_count.index, 'amount':survey_count.values})

    df_survey_gcm = pd.merge(df_survey_counted,
                df_cat_gcm[['resource','name', 'degree','eigenvector','pagerank']],
                #df_cat_gcm[['resource', 'degree','eigenvector','pagerank']],
                on='resource', 
                how='left')
    
    # Drop rows that were not matched with the category
    nan_rows = df_survey_gcm.shape[0] - df_survey_gcm.dropna().shape[0]    
    print("Rows with NaN values dropped: {}".format(nan_rows) + " from {}".format(df_survey_gcm.shape[0]) + " => {}% ".format(round((nan_rows/df_survey_gcm.shape[0])*100, 2))) #return amount of rows containing NaN values
    df_survey_gcm = df_survey_gcm.dropna()
    
    # Count how often an answer was given in the survey and save into new dataframe
    #df_survey_gcm['survey_mentions'] = df_survey_gcm['resource'].value_counts(normalize=True) #normalize values
    #df_survey_counted = pd.DataFrame({'resource':survey_count.index, 'amount':survey_count.values})
    
    # scale and normalize the survey data (same as scaler function)
    #to_scale = StandardScaler()
    #min_max_scaler = preprocessing.MinMaxScaler()
    #df_survey_gcm[['amount']] = min_max_scaler.fit_transform(df_survey_gcm[['amount']])
    
    # Calculates a normalised rank for the answer distribution per category
    df_survey_gcm['amount_rank'] = df_survey_gcm['amount']/np.sum(df_survey_gcm['amount'])

    
    print("Dataframe merged with shape: {}".format(df_survey_gcm.shape))
    
    return df_survey_gcm
```

```python
# Function for Linear Regression
  # takes the dataframe prepared by prepSurGCMLinReg as input
  # LinReg for Eigenvector and PageRank (because Degree Centrality and PR are to similar -> see correlation coeffizent -> overfitting)
    
def linReg(prepared_data_frame, gcm='ecpr'):
    helper_df = prepared_data_frame.copy()
    
    if gcm == 'pr':
        X = pd.DataFrame({"pagerank": helper_df.pagerank})
        print("PageRank Regression:")
    elif gcm == 'ec':
        X = pd.DataFrame({"eigenvector": helper_df.eigenvector})
        print("Eigenvector Regression:")
    elif gcm == "ecpr":
        X = pd.DataFrame({"eigenvector": helper_df.eigenvector, "pagerank": helper_df.pagerank})
        print("PageRank & Eigenvector Regression:")
   
    X = sm.add_constant(X, prepend=False)
    Y = helper_df.amount_rank # normalized survey results

    # Fit and summarize OLS model
    mod = sm.OLS(endog = Y, exog = X)

    res = mod.fit()
    robustreg = res.get_robustcov_results(cov_type='HC0')
    # HC0: heteroscedasticity robust covariance
    # https://www.statsmodels.org/devel/generated/statsmodels.regression.linear_model.OLSResults.get_robustcov_results.html

    #print(res.summary())
    print(robustreg.summary())
```

```python
def plotLinReg(prepared_data_frame, cat_name, gcm_string, number_of_annotations=5):
    label_nr = number_of_annotations
    gcm_name = gcm_string
    helper_df = prepared_data_frame.copy()
    save_name = cat_name.replace(" ","_")
    model = linear_model.LinearRegression()
    
    gcm_array = helper_df[gcm_name].to_numpy()
    gcm_array = gcm_array.reshape(-1, 1)
    sv_array = helper_df['amount_rank'].to_numpy()
    sv_array = sv_array.reshape(-1, 1)

    model.fit(sv_array, gcm_array) # fitting the model

    #print("The coefficient is:", model.coef_)
    #print("The intercept is:",model.intercept_)
    print(gcm_name)


    #Plot the data
    sns.scatterplot(data=helper_df, x="amount_rank", y=gcm_name, size=gcm_name, legend=False, hue=gcm_name, palette="deep")

    #Annotate data points
    labels = helper_df['name']
    # Loop through the data points
    
    if gcm_name == 'degree':
        for i, label in enumerate (labels):
            plt.annotate(label, (helper_df.amount_rank[i], helper_df.degree[i]))
            if i == label_nr:
                break
        y_pred = model.predict(sv_array)
        plt.plot(sv_array, y_pred, color = "r")
    elif gcm_name == 'eigenvector':
        for i, label in enumerate (labels):
            plt.annotate(label, (helper_df.amount_rank[i], helper_df.eigenvector[i]))
            if i == label_nr:
                break
        y_pred = model.predict(sv_array)
        plt.plot(sv_array, y_pred, color = "g")
    else:
        for i, label in enumerate (labels):
            plt.annotate(label, (helper_df.amount_rank[i], helper_df.pagerank[i]))
            if i == label_nr:
                break
        y_pred = model.predict(sv_array)
        plt.plot(sv_array, y_pred, color = "b")

    #plt.title('Scatter plot with EU capital cities and their PageRank', size=18)
    plt.xlabel('Mentions in survey', fontsize=14,fontweight='bold', labelpad=15)
    plt.ylabel(gcm_name.capitalize()+' centr. in DBpedia', fontsize=14,fontweight='bold', labelpad=15)
    plt.tick_params(labelsize=14)
    plt.title("Linear Regr. for {}".format(cat_name), size=15)


    plt.savefig("../plots/plots_linearRegression/"+save_name+"_LinReg_"+gcm_name+".png", bbox_inches="tight")
    plt.show()
```
